```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Gaussian window
def gaussian_window(t, t_0, sigma):
    return np.exp(-0.5 * ((t - t_0) / sigma) ** 2)

# Generate toy data
t = np.linspace(0, 10, 1000)  # Time values
t_0 = 8  # Center of the window
sigma = 0.7  # Width of the window
v = 2.5  # Velocity at time t_0
k = 0.0001  # Normalization constant

# Generate vertical acceleration data (toy data)
zacc = np.sin(t) * 2  # You can replace this with your actual acceleration data

# Calculate surface roughness using the provided formula
surface_roughness = k / v * np.sum(gaussian_window(t, t_0, sigma) * zacc * t**2)

# Plot the toy data and Gaussian window
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, zacc, label='Vertical Acceleration (zacc)')
plt.title('Toy Data: Vertical Acceleration')
plt.xlabel('Time (t)')
plt.ylabel('Acceleration')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, gaussian_window(t, t_0, sigma), label='Gaussian Window')
plt.title('Gaussian Window Function')
plt.xlabel('Time (t)')
plt.ylabel('Window Value')
plt.legend()

plt.tight_layout()
plt.show()

# Print the calculated surface roughness
print(f"Surface Roughness: {surface_roughness}")
```
